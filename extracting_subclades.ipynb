```python
# some miscellaneous details
sample_size = 10_000
seed        = 0
array_len   = 255
n_arrays    = 100
```

# Getting sequence ID set

```python
random.seed(seed)
pattern = os.path.join(raw_dir,"*.fa.gz")
for fasta_path in sorted(glob(pattern)):
    dropout    = os.path.basename(fasta_path).replace(".fa.gz","").split("_")[2]
    if dropout in ["d60","d70","d80","d90"]: continue
    fasta_dict = { seq_id:seq for seq_id,seq in Fasta(fasta_path,build_index=False) }
    ids        = list(fasta_dict.keys())
    for i in tqdm(range(3),desc=dropout):
        sample_label = f"sample{i+1}"
        sampled_ids  = random.sample(ids,sample_size)
        id_path      = os.path.join(id_dir,f"PRESUMEBE10k.{dropout}.{sample_label}.ids.txt.gz")
        with gzip.open(id_path,"wt") as handle:
            handle.write("\n".join(sampled_ids)+"\n")
```

# For each subsampled ID set, extract the corresponding sequences from the complete 1M sequence FASTA file

* Just using subprocess to keep everything in this script
* These subsampled FASTA files will correspond to the "complete" (i.e., all 100 barcode arrays per cell) dataset

```python
for id_path in tqdm(sorted(glob(os.path.join(id_dir,"*.ids.txt.gz")))):
    label   = os.path.basename(id_path).replace(".ids.txt.gz","")
    dropout = label.split(".")[1]
    complete_fasta_path   = os.path.join(raw_dir,f"tip1000000_seed91743010_{dropout}.fa.gz")
    subsampled_fasta_path = os.path.join(fasta_dir,f"{label}.complete.fa.gz")
    command = f"seqkit grep -f {id_path} {complete_fasta_path} | gzip > {subsampled_fasta_path}"
    process = subprocess.run(command,shell=True,capture_output=True)
    if process.returncode != 0:
        print(process.stdout.decode("utf-8"))
        print(process.stderr.decode("utf-8"))
        raise Exception()
```

# Extract FASTA for each individual barcode array sequence

* Skip subsequences if all gap ("-") characters

```python
for fasta_path in tqdm(sorted(glob(os.path.join(fasta_dir,"*.complete.fa.gz")))):
    ids  = []
    seqs = []
    for seq_id,seq in Fasta(fasta_path,build_index=False):
        ids.append(seq_id)
        seqs.append(list(seq))
    seqs = np.array(seqs)
    n,m  = seqs.shape
    for i,j in enumerate(range(0,m,array_len)):
        barcode = f"barcode{i+1:0>3}"
        outpath = fasta_path.replace(".complete.fa.gz",f".{barcode}.fa.gz")
        subseqs = [ "".join(seq) for seq in seqs[:,j:j+array_len] ]
        with gzip.open(outpath,"wt") as handle:
            for seq_id,subseq in zip(ids,subseqs):
                assert len(subseq) == array_len
                if "-"*array_len == subseq: continue # the sequence is all gaps (i.e., dropout)
                handle.write(f">{seq_id}\n{subseq}\n")
```
